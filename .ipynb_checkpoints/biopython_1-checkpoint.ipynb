```python
def quick_plot_all( data ):
    x = {}
    y = {}
    z = {}
    for models in data: 
        x[models] = []
        y[models] = []
        z[models] = []
        for i in data[models]['coord']: 
            x[models].append(i[0])
            y[models].append(i[1])
            z[models].append(i[2])
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(projection = '3d')
    for models in x: 
        plt.plot(x[models],y[models],z[models], alpha=0.8)
    # ax.scatter(x, y, z, c='skyblue', s=10)
    ax.view_init()
    plt.show()
    return 
def quick_plot( coordinate ):
    x = []
    y = []
    z = []
    for i in coordinate: 
        x.append(i[0])
        y.append(i[1])
        z.append(i[2])

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(projection = '3d')
    plt.plot(x,y,z)
    ax.scatter(x, y, z, s=60)
    ax.view_init()
    plt.show()
    return 
```

```python
pdb_files = os.listdir("./pdb")
pdb_files
```

```python
# opens and reads .pdb files and outputs 3 lists of atom, position coordinate and resid number
data = {}
for i in pdb_files:
    name = i.split("_")[1].split(".")[0]
    atom, coord, resid = rmsd.get_coordinates_pdb("./pdb/" + i)
    data[name] = {}
    data[name]["atom"] = atom
    data[name]["coord"] = coord
    data[name]["resid"] = resid
```

```python
# quick plot to visualize protein location in the 3d domain 
quick_plot_all(data)
```

```python
# remove "H" from models, since some models doesn't have it and it may throw off alignment 
for models in data:     
    H_atom = ["H" in i for i in data[models]['atom']]
    non_H_atom = np.invert(H_atom)
    data[models]['atom'] = data[models]['atom'][non_H_atom]
    data[models]['coord'] = data[models]['coord'][non_H_atom]
    data[models]['resid'] = data[models]['resid'][non_H_atom]
```

```python
# create a backbone_data for measuring center, RMSD and rotation 
backbone = copy.deepcopy(data)
# bb_atoms = ["N","CA","CB","C","O"]
bb_atoms = []
for models in backbone: 
    N_atom = ["N" == i for i in backbone[models]['atom']]
    CA_atom = ["CA" == i for i in backbone[models]['atom']]
    CB_atom = ["CB" == i for i in backbone[models]['atom']]
    C_atom = ["C" == i for i in backbone[models]['atom']]
    O_atom = ["O" == i for i in backbone[models]['atom']]
    # intersect all the True index 
    temp = np.logical_or(N_atom,CA_atom)
    temp = np.logical_or(temp,CB_atom)
    temp = np.logical_or(temp,C_atom)
    temp = np.logical_or(temp,O_atom)
    for i in backbone[models]:
        backbone[models][i] = backbone[models][i][temp]
    
```

```python
# take a peak at atom orders for backbones 
for models in backbone:
    print( models , backbone[models]['atom'][0:10])
```

```python
# reorder the atoms, coord and resid so that atom orders are ('C' 'CA' 'CB' 'N' 'O' ...)
# sorted by alphabatical order for simplicity 
for models in backbone: 
    orig_index = []
    reordered_index = []
    for resid in np.unique(backbone[models]['resid']):
        (orig_index,) = np.where(backbone[models]['resid'] == resid)
        reordered_index = np.argsort(backbone[models]['atom'][orig_index]) + np.min(orig_index)
        backbone[models]['atom'][orig_index] = copy.deepcopy(backbone[models]['atom'][reordered_index])
        backbone[models]['coord'][orig_index] = copy.deepcopy(backbone[models]['coord'][reordered_index])
```

```python
# visualizaiton after removing H atoms
quick_plot_all(backbone)
```

```python
# Manually subset the resid numbers for RMSD calculation, since calculating RMSD requires the atom numbers to be same 
# typically omit the ends (resid 25 - 285)
for models in backbone: 
    keep_resid = np.where((backbone[models]['resid'] <= 285) & (backbone[models]['resid'] >= 25 ))
    backbone[models]['atom'] = backbone[models]['atom'][keep_resid]
    backbone[models]['coord'] = backbone[models]['coord'][keep_resid]
    backbone[models]['resid'] = backbone[models]['resid'][keep_resid]
```

```python
# visualizaiton after centering
quick_plot_all(backbone)
print("RMSD backbone pontcy - cryoEM : ",rmsd.rmsd(backbone['pontcy']['coord'], backbone['cryoEM']['coord']))
print("RMSD backbone alphafold - cryoEM : ",rmsd.rmsd(backbone['alphafold']['coord'], backbone['cryoEM']['coord']))
```

```python
# Center models to center by substracting coordinates by centroid 
for models in backbone: 
    center = rmsd.centroid(backbone[models]['coord'])
    backbone[models]['coord'] -= center
```

```python
# visualizaiton after removing ends 
quick_plot_all(backbone)
print("RMSD backbone pontcy - cryoEM : ",rmsd.rmsd(backbone['pontcy']['coord'], backbone['cryoEM']['coord']))
print("RMSD backbone alphafold - cryoEM : ",rmsd.rmsd(backbone['alphafold']['coord'], backbone['cryoEM']['coord']))
```

```python
# Get rotation matrix, can only compare x to y so pick a module to align to 
align = 'cryoEM'
for models in backbone: 
    if models == align: 
        continue 
    backbone[models]['coord'] =  rmsd.kabsch_fit(backbone[models]['coord'], backbone[align]['coord']) 
```

```python
# visualizaiton after rotation alignment
quick_plot_all(backbone)
print("RMSD backbone pontcy - cryoEM : ",rmsd.rmsd(backbone['pontcy']['coord'], backbone['cryoEM']['coord']))
print("RMSD backbone alphafold - cryoEM : ",rmsd.rmsd(backbone['alphafold']['coord'], backbone['cryoEM']['coord']))
```

```python
# Now that the structures are aligned find the RMSD for every resid in respect to cryoEM 
align = 'cryoEM'
rmsd_data = {}
rmsd_resid = np.unique(backbone['cryoEM']['resid'])
for models in backbone: 
    if models == align: 
        continue 
    rmsd_data[models] = []
    for resid in np.unique(backbone[models]['resid']):
        resid_index = np.where(backbone[models]['resid'] == resid)
        temp = rmsd.rmsd(backbone['cryoEM']['coord'][resid_index], backbone['alphafold']['coord'][resid_index])
        rmsd_data[models].append(temp)
```

```python
# export rmsd data frame as a csv to plot in R 
rmsd_data = pd.DataFrame(rmsd_data, rmsd_resid)
rmsd_data.to_csv("./rmsd_data.csv", index=True)
```
